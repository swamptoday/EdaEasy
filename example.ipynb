# Example of how to use EDAReport for regression and classification problems

```python
import pandas as pd
from importlib import reload
```

```python
from edaeasy.report.edareport import EDAReport
```

## EDA for regression (target variable is numerical)

```python
df = pd.read_csv('example/chocolate_bars.csv')
```

```python
eda = EDAReport(df, problem_type='regression', target='rating', path='example/image_keeper')
```

### Perform basic analysis 
* includes information, head,shape, data types, duplicates and unique values

```python
eda.perform_basic()
```

### Perform target analysis
* for regression problem: description statistics and descriptive plot, distribution of target variable (boxplot, violinplot and histplot), outliers - amount and their percentage

* for classificaion problem: unique values and their amount, distribution with percentage (barplot)

```python
eda.perform_target()
```

### Perform univariate analysis
* for numerical features: descriptive statistics and description plot, distribution (boxplot, violinplot and histplot), outliers (amount and percentage), missing values (heatmap)

* for categorical features: ditribution with percentage (barplot), missing values (heatmap)

```python
eda.perform_univariate()
```

### Perform multivariate analysis
* for regression problem: correlation matrix WITH target variable (heatmap), high correlation pairs relations (jointplot), MI score (barplot), relations between all numerical features (pairplot), distribution of categorical features between target (boxplot)

* for classification problem: correlation matrix WITHOUT target variable (heatmap), high correlation pairs relations (jointplot), distribution of numerical features between target (boxplot), distribution of categorical features between target (barplot)

```python
eda.perform_multivariate()
```

## EDA for classification problem (target variable is categorical)

```python
df = pd.read_csv('example/stroke_data.csv')
```

```python
eda_class = EDAReport(df, problem_type='classification', target='stroke', path='example/image_keeper')
```
